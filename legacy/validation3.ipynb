```python
fuelID = 'Oltenia'
moist = fu.moisture(fuelID)
T = 580 + 273.15 # K
P = 1E6 # Pa
O2_ratio = 0.35
print(moist)
```

```python
report, feed, outlet, fuelMix = g.gasifier(fuelID, fuelMass=1.0, moist=moist, T=T, P=ct.one_atm, 
                air=1.957, stm=0.0, airType='air', stmType='SR', C=1.0, CH4=0.0, isot=True, heatLoss=0.0,
                species=['C(gr)','N2','O2','H2','CO','CH4','CO2','H2O'])
```

```python
stoich_ratio = fs.stoichO2(fuelMix)
stoich_O2_moles = stoich_ratio*sum(fuelMix.species_moles)
O2_moles = O2_ratio*stoich_O2_moles
O2_mass = O2_moles*pp.Mw['O2']
print(O2_mass)
```

```python
report, feed, outlet, fuelMix = g.gasifier(fuelID, fuelMass=1.0, moist=moist, T=T, P=P, 
                air=O2_mass, stm=0.0, airType='O2', stmType='SR', C=1.0, CH4=0.0, isot=True, heatLoss=0.0,
                species=['C(gr)','N2','O2','H2','CO','CH4','CO2','H2O'])
```

```python
for sp in ['CO', 'CO2', 'CH4', 'H2', 'N2', 'H2O']:
    print(f'{sp} = ', report[sp])
print(report['HHV'])
print(report['Y'])
```
